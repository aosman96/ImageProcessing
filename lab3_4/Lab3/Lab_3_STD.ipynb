```python
'''
Averaging Filters
(1)
'''
def calculate_median(i,j,image):
    A= np.array([image[i,j],image[i+1,j],image[i-1,j],image[i+1,j+1],image[i+1,j-1],image[i,j+1],image[i,j-1],image[i-1,j+1],image[i-1,j-1]])
    A= np.sort(A)
    return A[4]
 
def median_filter(img):
    shape = img.shape
    new_img = np.zeros([shape[0],shape[1]])
    for i in range(shape[0]):
        for j in range(shape[1]):
            if (i==0 or i==shape[0]-1 or j==0 or j==shape[1]-1 ):
                new_img[i,j] = img[i,j]
                continue
            new_img[i,j]=calculate_median(i,j,img)
    return new_img        
    
```

```python
coffee = io.imread('coffee.jpeg')
coffee_grayscale = rgb2gray(coffee)
noise_coffe = random_noise(coffee_grayscale,mode='s&p',amount=0.05)
med_img = median_filter(coffee_grayscale)
skmedian = median(noise_coffe)
gssimg1 = gaussian(noise_coffe,sigma=0.5)
gssimg2 = gaussian(noise_coffe,sigma=8)
gssimg3 = gaussian(noise_coffe,sigma=0.2)
```
