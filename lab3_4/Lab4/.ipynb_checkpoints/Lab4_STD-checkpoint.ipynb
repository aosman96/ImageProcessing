```python
import skimage.io as io
import matplotlib.pyplot as plt
import numpy as np

# Show the figures / plots inside the notebook
%matplotlib inline
from skimage.color import rgb2gray

def show_images(images,titles=None):
    #This function is used to show image(s) with titles by sending an array of images and an array of associated titles.
    # images[0] will be drawn with the title titles[0] if exists
    # You aren't required to understand this function, use it as-is.
    n_ims = len(images)
    if titles is None: titles = ['(%d)' % i for i in range(1,n_ims + 1)]
    fig = plt.figure()
    n = 1
    for image,title in zip(images,titles):
        a = fig.add_subplot(1,n_ims,n)
        if image.ndim == 2: 
            plt.gray()
        plt.imshow(image)
        a.set_title(title)
        n += 1
    fig.set_size_inches(np.array(fig.get_size_inches()) * n_ims)
    plt.show()
```

```python
'''
1
Negative Transformation
'''
def NegativeTransformation(img):
    shape  = img.shape
    new_img = np.ones([shape[0],shape[1]])
    #for i in range(shape[0]):
     #   for j in range(shape[1]):
      #      new_img[i,j] = 255 - img[i,j]
    new_img = 255 - img
    return new_img
pic1 = io.imread('imgs/Picture1.png')
pic1_gray = rgb2gray(pic1)
pic1_neg = NegativeTransformation(pic1_gray)
show_images([pic1,pic1_gray,pic1_neg],['orignal image','gray image','pic1 negative'])
```

```python
'''
2
Contrast Enhancement
'''
def Contrast_enhancement(img):
    shape = img.shape
    #bool_img  = img > 100
    new_img = np.ones([shape[0],shape[1]])
    #new_img = img 
    #print( bool_img )
    #new_img[ bool_img] = 2 * img[bool_img] +50
    for i in range(shape[0]):
        for j in range(shape[1]): 
            if img[i,j]>100:
                new_img[i,j] = img[i,j] *2 +50  
            else:
                new_img[i,j] = img[i,j]
                
    return new_img

pic2 = io.imread('imgs/Picture2.png')
pic2_gray = rgb2gray(pic2)
pic2_contrast_enhancement = Contrast_enhancement(pic2_gray)
show_images([pic2,pic2_gray,pic2_contrast_enhancement],['orignal image','gray image','Contrast_enhancement'])
```

```python
'''
3
Gamma Correction
'''
def Gamma_correction(img,c,y):
    new_image = np.ones([img.shape[0],img.shape[1]])
    new_image = c * np.power(img,y)
    return new_image
pic2 = io.imread('imgs/tire.tif')
pic2_gray = rgb2gray(pic2)
pic2_Gamma_correction1 = Gamma_correction(pic2_gray,1,3)
pic2_Gamma_correction2 = Gamma_correction(pic2_gray,1,0.5)
show_images([pic2,pic2_Gamma_correction1,pic2_Gamma_correction2],['orignal image','pic2_Gamma_correction 3','pic2_Gamma_correction 0.5'])
ti = histogram(pic2_gray)
print(ti)
```

```python
'''
4 Histogram Eq.
Note: Histogram function of skimage returns only present intensity values not all 255
You can use showHist function from commonfunctions file
'''

def getImageWithHist(img,nbins=256):
    H = np.zeros(nbins)
    new_img = np.zeros([img.shape[0],img.shape[1]])
    histimg1,histimg2 = histogram(img,nbins)
    #print(len(histimg2))
    for i in range(len(histimg2)):
       # print(int(histimg2[i] *255))
        H[int(histimg2[i])] = histimg1[i]
        
    cum = np.cumsum(H)
    #new_img =(nbins-1)*cum[img]/( img.shape[0] * img.shape[1])
   # print(img)
    #print(H)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            new_img[i,j] = (nbins-1)*cum[img[i,j]]/( img.shape[0] * img.shape[1])
                
    histimg = (histimg1,histimg2)
   # print(new_img.shape)
    return new_img , histimg
pic3 = io.imread('imgs/tire.tif')
pic3_gray = rgb2gray(pic3)
pics3_eqhist , pic3_hist = getImageWithHist(pic3,256)

show_images([pic3,pic3_gray,pics3_eqhist],['orignal image','gray image ',' after equ' ])
showHist(pic3)
showHist(pic3_gray)
showHist(pics3_eqhist)
```
